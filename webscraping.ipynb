WebScrapping

```python
import numpy as np
import pandas as pd

import requests
from bs4 import BeautifulSoup
```

```python
URL = 'https://timely-sunshine-e821b3.netlify.app/'

```

```python
page = requests.get(URL)
```

```python
page.status_code
```

```python
htmlcode = page.text
soup = BeautifulSoup(htmlcode)
```

```python
htmlcode
```

```python
content = soup.find('div', attrs = {'class':'main'})
text = content.text.strip()
r1 = text.split('\n')

r2 =[]
for i in r1:
  if i != '':
    r2.append(i)
r3 = []
for i in r2:
  r3.append(i.strip())

r4 = []
for i in r3:
  if i!= '':
    r4.append(i)
#print(r4)
names = []
prices = []
for i in range(0,len(r4),2):
  names.append(r4[i])
  prices.append(r4[i+1])
print(names)
print(prices)
```

```python
df = pd.DataFrame({'product_name':names, 'price':prices})
```
